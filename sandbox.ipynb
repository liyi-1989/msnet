```r
onebump=function(t,px){
  (floor(0.2*px) < t && t < floor(0.4*px))*(1-abs((floor(0.3*px) - t)/floor(0.1*px)))
}
```

```r
px=30
x=rep(0,px)
for(i in 1:px){
    x[i]=onebump(t=i,px=px)
}
```
